**Import**

```python
import pandas as pd
```

**Load**

```python
stock_file = '/home/tiennv/FPT/Finance_Forecast/Stock_Predictor/data/external/stock/ver_1/sp500_stocks.csv'
```

```python
stock_df = pd.read_csv(stock_file, encoding='UTF-8')
stock_df
```

**Preprocess**

```python
stock_df['Date'] = pd.to_datetime(stock_df['Date'])
stock_df
```

```python
stock_df = stock_df.sort_values(by='Date')
stock_df
```

**Filter Sector**

```python
sector = 'Technology'
companies_file = ''
```

```python
companies_df = pd.read_csv(companies_file, encoding='UTF-8')
companies_df
```

```python
symbols = companies_df[companies_df['Sector'] == sector]['Symbol'].unique()
symbols
```

```python
len(symbols)
```

```python
stock_df = stock_df[stock_df['Symbol'].isin(symbols)]
stock_df
```

**Choose Date Range**

```python
start_date = '2010-01-01'
end_date = '2024-04-30'
```

```python
stock_df = stock_df[(stock_df['Date'] >= start_date) & (stock_df['Date'] <= end_date)]
stock_df
```

**Handle NaN values**

```python
stock_df = stock_df.dropna()
stock_df
```

**Check compatibility with the Stock Embedding Model**

```python
scaler_file = '/home/tiennv/FPT/Finance_Forecast/Stock_Predictor/models/trained_models/stock_embedder/ver_2_1/scaler.pkl'
```

```python
import joblib
```

```python
scaler = joblib.load(scaler_file)
```

```python
stock_features = scaler.feature_names_in_
stock_features
```

```python
flag = False

for feature in stock_features:
    if feature not in stock_df.columns:
        print(f'{feature} not found in stock_df')
        flag = True

if flag:
    print('Columns in stock_df are: ', stock_df.columns)
    raise Exception()
```
